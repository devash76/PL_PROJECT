```python
api = tweepy.API(auth)
name = "economictimes"
tweetCount = 200
results = api.user_timeline(id=name, count=tweetCount)
```

```python
i=0
tweet_data={}
for tweet in results:
    print(tweet.text)
    print()
    print('Date and Time of Tweet:-')
    print(tweet.created_at) #date of tweet
    tweet_data[i]={"text":tweet.text,
                   "date":str(tweet.created_at),
                   "URL_ID":str(tweet.id),
                   "Likes":tweet.user.favourites_count,
                   "Retweet":tweet.retweet_count,
                   "Hashtags": tweet.entities['hashtags']}

    i+=1
    
    
print('Profile Name:-')
print(tweet.user.name) 
print()
print('Twitter ID:-')
print(tweet.user.screen_name)
print()
print('Bio:-')
print(tweet.user.description)
print()
print('Location:-')
print(tweet.user.location) 
print()
print('URL:-')
print(tweet.user.url) 
print()
print('Joined:-')
print(tweet.user.created_at) 
print()
print('Following:-')
print(tweet.user.friends_count)
print()
print('Followers:-')
print(tweet.user.followers_count) 
print()
print('Profile Image:-')
print(tweet.user.profile_image_url_https) 
print()
print('Coverpage:-')
print(tweet.user.profile_banner_url) 
print()
print('Verification:-')
print(tweet.user.verified) 
print()
```
