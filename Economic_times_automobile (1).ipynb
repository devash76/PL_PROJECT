```python
import requests
from bs4 import BeautifulSoup
import json
links=[]
subheadlines=[]
headlines=[]
images=[]
date=[]
authors=[]
text_format=[]
page = requests.get("https://economictimes.indiatimes.com/industry/auto/auto-news/articlelist/msid-64829342,contenttype-a.cms")
soup = BeautifulSoup(page.text, "html.parser")

article_list=soup.find_all(class_="eachStory")
for article in article_list:
    link= article.find('a')
    req = 'https://economictimes.indiatimes.com'+ link.get('href')
    links.append(req)
    newspage = requests.get(req)
    soup = BeautifulSoup(newspage.text,"html.parser")
    datetime=soup.find(class_="publish_on flt")
    date.append(datetime.text)
    fonts = soup.find_all(class_="Normal")
    for i in range(len(fonts)):
        text_format.append(fonts[i].text)
    writer=soup.find(class_="publisher flt")
    authors.append(writer.text)
    pictures = soup.find_all("figure")
    for picture in pictures:
        l=picture.find("img")
        k=l.get("src")
        images.append(k)
   
    t=article.find('p')
    subheadlines.append(t.text)
    h=article.find('h3')
    headlines.append(h.text)
    
s1=json.dumps(links,indent=4)
s2=json.dumps(subheadlines,indent=4)
s3=json.dumps(headlines,indent=4)
s4=json.dumps(images,indent=4)
s5=json.dumps(date,indent=4)
s6=json.dumps(authors,indent=4)
s7=json.dumps(text_format,indent=4)
with open('links.json','w') as f1:
    json.dump(s1,f1)
with open('subheadlines.json','w') as f2:
    json.dump(s2,f2)
with open('headlines.json','w') as f3:
    json.dump(s3,f3)
with open('images.json','w') as f4:
    json.dump(s4,f4)
with open('date.json','w') as f5:
    json.dump(s5,f5)
with open('authors.json','w') as f6:
    json.dump(s6,f6)
with open('text_format.json','w') as f7:
    json.dump(s7,f7)
    print(s1)
    print(s2)
    print(s3)
    print(s4)
    print(s5)
    print(s6)
    print(s7)






    

    
```
