```python
import requests
from bs4 import BeautifulSoup
links=[]
subheadlines=[]
headlines=[]
images=[]
date=[]
authors=[]
text_format=[]
page = requests.get("https://economictimes.indiatimes.com/industry/auto/auto-news/articlelist/msid-64829342,contenttype-a.cms")
soup = BeautifulSoup(page.text, "html.parser")

article_list=soup.find_all(class_="eachStory")
for article in article_list:
    link= article.find('a')
    req = 'https://economictimes.indiatimes.com'+ link.get('href')
    links.append(req)
    newspage = requests.get(req)
    soup = BeautifulSoup(newspage.text,"html.parser")
    datetime=soup.find(class_="publish_on flt")
    date.append(datetime.text)
    fonts = soup.find_all(class_="Normal")
    for i in range(len(fonts)):
        text_format.append(fonts[i].text)
    writer=soup.find(class_="publisher flt")
    authors.append(writer.text)
    pictures = soup.find_all("figure")
    for picture in pictures:
        l=picture.find("img")
        k=l.get("src")
        images.append(k)
   
    t=article.find('p')
    subheadlines.append(t.text)
    h=article.find('h3')
    headlines.append(h.text)
import os,json
i=1
Dict ={}
ulist = open("linkslist.txt",'w')
for link in links:
    ulist.write(link+'\n')
    Dict["Page "+str(i)]=link
    i+=1
ulist.close()

File = json.dumps(Dict)
jso = open("Links_list.json",'w')
jso.write(File)
jso.close()
    
i=1
Dict ={}
ulist = open("subheadings.txt",'w')
for sub in subheadlines:
    ulist.write(sub+'\n')
    Dict["Page "+str(i)]=sub
    i+=1
ulist.close()

File = json.dumps(Dict)
jso = open("subheadlines.json",'w')
jso.write(File)
jso.close()


i=1
Dict ={}
ulist = open("Headlines.txt",'w')
for head in headlines:
    ulist.write(head+'\n')
    Dict["Page "+str(i)]=head
    i+=1
ulist.close()

File = json.dumps(Dict)
jso = open("headlines.json",'w')
jso.write(File)
jso.close()


i=1
Dict ={}
ulist = open("images.txt",'w')
for image in images:
    ulist.write(image+'\n')
    Dict["Page"+str(i)]=image
    i+=1
ulist.close()

File = json.dumps(Dict)
jso = open("images.json",'w')
jso.write(File)
jso.close()

i=1
Dict ={}
ulist = open("date.txt",'w')
for dt in date:
    ulist.write(dt+'\n')
    Dict["Page"+str(i)]=dt
    i+=1
ulist.close()

File = json.dumps(Dict)
jso = open("dates.json",'w')
jso.write(File)
jso.close()

i=1
Dict ={}
ulist = open("authors.txt",'w')
for author in authors:
    ulist.write(author+'\n')
    Dict["Page"+str(i)]=author
    i+=1
ulist.close()

File = json.dumps(Dict)
jso = open("authors.json",'w')
jso.write(File)
jso.close()

i=1
Dict ={}
ulist = open("text_format.txt",'w')
for text in text_format:
    ulist.write(text+'\n')
    Dict["Page"+str(i)]=text
    i+=1
ulist.close()

File = json.dumps(Dict)
jso = open("text.json",'w')
jso.write(File)
jso.close()
```
